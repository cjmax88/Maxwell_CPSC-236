````python
import csv

def read_monthly_sales(file_name):
    monthly_sales = {}
    with open(file_name, 'r', newline='') as file:
        reader = csv.reader(file)
        for row in reader:
            month, sales = row
            monthly_sales[month] = int(sales)
    return monthly_sales

def write_monthly_sales(monthly_sales, file_name):
    with open(file_name, 'w', newline='') as file:
        writer = csv.writer(file)
        for month, sales in monthly_sales.items():
            writer.writerow([month, sales])

def calculate_summary(monthly_sales):
    yearly_total = sum(monthly_sales.values())
    monthly_average = yearly_total / len(monthly_sales)
    return yearly_total, round(monthly_average, 2)

def main():
    monthly_sales_file = 'monthly_sales.csv'
    monthly_sales = read_monthly_sales(monthly_sales_file)
    
    while True:
        print("\nMonthly Sales program\n")
        print("COMMAND MENU")
        print("monthly - View monthly sales")
        print("yearly  - View yearly summary")
        print("edit    - Edit sales for a month")
        print("exit    - Exit program")
        
        command = input("\nCommand: ").lower()
        
        if command == "monthly":
            print("Monthly Sales:")
            for month, sales in monthly_sales.items():
                print(month, "-", sales)
        
        elif command == "yearly":
            yearly_total, monthly_average = calculate_summary(monthly_sales)
            print("Yearly total: ", yearly_total)
            print("Monthly average: ", monthly_average)
        
        elif command == "edit":
            month = input("Three-letter Month: ").capitalize()[:3]
            if month in monthly_sales:
                try:
                    sales = int(input("Sales Amount: "))
                    monthly_sales[month] = sales
                    write_monthly_sales(monthly_sales, monthly_sales_file)
                    print("Sales amount for ",month,"was modified.")
                except ValueError:
                    print("Invalid input. Please enter a valid sales amount.")
            else:
                print("Invalid month abbreviation. Please try again.")
        
        elif command == "exit":
            print("Bye!")
            break
        
        else:
            print("Invalid command. Please try again.")

if __name__ == "__main__":
    main()
````
